```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# Load CSV file
csv_path = os.path.join("Resources", "purchase_data.csv")

#set float format
pd.options.display.float_format = '${:,.2f}'.format

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(csv_path)
```

## Player Count

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#perform the summary calculations
unique_items=len(purchase_data["Item Name"].unique())
avgPrice=purchase_data["Price"].mean()
nPurchases=purchase_data["Purchase ID"].count()
TotalRevenue=purchase_data["Price"].sum()

#create the summary dataframe
summary_df=pd.DataFrame({"Number of Unique Items":[unique_items], "Average Price": [avgPrice], 
                        "Number of Purchases": [nPurchases], "Total Revenue":[TotalRevenue]})

#display the dataframe
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
players_list=purchase_data.drop_duplicates(subset="SN", keep="first")

nPlayers=players_list["Gender"].count()

#count of players in each category
Male_players=players_list["Gender"].value_counts()["Male"]
Female_players=players_list["Gender"].value_counts()["Female"]
Other_players=players_list["Gender"].value_counts()["Other / Non-Disclosed"]

#percentage of players in each category
perMale= (Male_players/nPlayers)*100
perFemale= (Female_players/nPlayers)*100
perOther= (Other_players/nPlayers)*100

#create the dataframe
GenDemo_df=pd.DataFrame({"": ["Male", "Female", "Other/Non-Disclosed"],
                        "Total Count": [Male_players, Female_players, Other_players],
                        "Percentage of Players": [perMale, perFemale, perOther]})

#Format percentages
GenDemo_df["Percentage of Players"]=GenDemo_df["Percentage of Players"].map("{:.2f}%".format)

GenDemo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_groups=purchase_data.groupby(["Gender"])

#Calculate courchase count, aveage purchase, and average price
nPurchase=gender_groups["Purchase ID"].count()
avgPrice=gender_groups["Price"].mean()
sumRevenue=gender_groups["Price"].sum()

#Find total number of payers per group to find avegage purchase per group
players_list=purchase_data.drop_duplicates(subset="SN", keep="first")
grouped_players = players_list.groupby(["Gender"])
nPlayers=grouped_players["Gender"].count()

AvgPurchase=sumRevenue/nPlayers

#create the summary dataframe
gender_summary_df=pd.DataFrame({"Purchase Count":nPurchase, "Average Purchase Price": avgPrice, 
                                "Total Purchase Value":sumRevenue,"Avg Total Purchase per Person":AvgPurchase})

#display the dataframe
gender_summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
players_list=purchase_data.drop_duplicates(subset="SN", keep="first")

Total_Players=players_list["SN"].count()

#create age bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

players_list["Age Group"] = pd.cut(players_list["Age"], bins, labels=group_names)
age_group_df=players_list.groupby(["Age Group"])

#count number of players in each bin and find the percentage
AgeCount = age_group_df["SN"].count()
perAgeGroup = (AgeCount / Total_Players) * 100

age_group_df=pd.DataFrame({"Total Count": AgeCount,
                         "Percentage of Players": perAgeGroup})

#format the percentage
age_group_df["Percentage of Players"]=age_group_df["Percentage of Players"].map("{:.2f}%".format)
age_group_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

#get the number of customers in each age group
players_list=purchase_data.drop_duplicates(subset="SN", keep="first")
age_data=purchase_data

Total_Players=players_list["SN"].count()

#create age bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

players_list["Age Group"] = pd.cut(players_list["Age"], bins, labels=group_names)
age_group_df=players_list.groupby(["Age Group"])

nPlayers = age_group_df["SN"].count()


#calculate the N purchases, Avg price, total revenue and average purchase per age group

age_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
age_group_df=age_data.groupby(["Age Group"])

nPurchase=age_group_df["Purchase ID"].count()
avgPrice=age_group_df["Price"].mean()
sumRevenue=age_group_df["Price"].sum()


AvgPurchase=sumRevenue/nPlayers

#create dataframe
age_group_summary_df=pd.DataFrame({"Purchase Count":nPurchase, "Average Purchase Price": avgPrice, 
                                "Total Purchase Value":sumRevenue,"Avg Total Purchase per Person":AvgPurchase})


age_group_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
customer_groups=purchase_data.groupby(["SN"])

#find purchase count, average purchase price and total purchase value
nPurchase_df = customer_groups.count()["Price"]
agePrice_df = customer_groups.mean()["Price"]
sumRevenue_df = customer_groups.sum()["Price"]


customer_df = pd.DataFrame({"Purchase Count": nPurchase_df,
                             "Average Purchase Price": agePrice_df,
                              "Total Purchase Value": sumRevenue_df})


# Sort table in descending order by purchase value
top_spenders_df = customer_df.sort_values("Total Purchase Value", ascending=False)
top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df=purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

#group by Item ID and Item Name
items_grouped_df=items_df.groupby(["Item ID", "Item Name"])

#find purchase count, item price, and total purchase value
nPurchases_df=items_grouped_df.count()["Price"]
sumRevenue_df=items_grouped_df.sum()["Price"]
price_df=items_grouped_df.mean()["Price"]

items_summary_df = pd.DataFrame({"Purchase Count":nPurchases_df,"Item Price":price_df, 
                                 "Total Purchase Value": sumRevenue_df})

item_Purchase_Count_sort = items_summary_df.sort_values("Purchase Count", ascending=False)

item_Purchase_Count_sort.head()
```
